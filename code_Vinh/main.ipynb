```python
from scipy.linalg import lu
from numpy.linalg import inv
import numpy as np

A = np.array([[0, 8, 2],
              [3, 5, 2],
              [6, 2, 8]])
b = np.array([[-7, 8, 26]]).T

P, L, U = lu(A)

# Ly = b => y = L^-1 b
y = np.dot(inv(L[[2,0,1], :]), b)


# Ux = y
x = np.dot(inv(U[[2,0,1], :]), y)

print(x)
```

PHƯƠNG PHÁP CHIA ĐÔI

```python

import numpy as np
from math import log, exp

def f(x):
    return 2*x + exp(x)   # tuy de se cho

TOL = 1e-3  # sai so yeu cau 10D => 1e-11
No = 100000   # so lan lap
a = -10  # can duoi
b = 10   # can tren
FA = f(a)  # gia tri tai f(a)
p = 1  # Nghiem ban dau, du doan = 1

result = p

for i in range(No):
    result = p = a + (b-a)/2
    FP = f(p)
    if FP == 0 or (b-a)/2 < TOL:
        result = p
        break
    elif FA * FP > 0:
        a = p
        FA = FP
    else:
        b = p

print(result)
```

PHƯƠNG PHÁP NEWTON

```python


import numpy as np
from math import log, exp, sqrt

def f(x):
    return x**3 - 2*x**2 - 5

def df(x):
    global f
    h = 1e-3
    return (f(x + h) - f(x)) / h

TOL = 1e-4   # sai so yeu cau 10D => 1e-11
No = 1000    # so lan lap
p = 5*np.random.rand(1)  # Nghiem ban dau random
prev_p = p

for i in range(No):
    p = p - f(p)/df(p)
    if abs(p - prev_p) < TOL:
        break
    prev_p = p

print(p)
```

```python
import numpy as np
from math import log, exp, sqrt

def f(x):
    return x**3 - 2*x**2 - 5

def df(x):
    global f
    h = 1e-3
    return (f(x + h) - f(x)) / h

TOL = 1e-4 # sai so yeu cau 10D => 1e-11
No = 1000  # so lan lap
p0 = 5*np.random.rand(1)  # Nghiem ban dau random
p1 = p0 + 10
q0 = f(p0)
q1 = f(p1)

for i in range(No):
    p = p1 - q1*(p1 - p0)/(q1 - q0)
    if abs(p - p1) < TOL:
        break
    p0, q0, p1, q1 = p1, q1, p, f(p)

print(p)
```

```python

import numpy as np
from math import log, exp, sqrt

def f(x):
    return x**3 - 2*x**2 - 5

def df(x):
    global f
    h = 1e-3
    return (f(x + h) - f(x)) / h

TOL = 1e-4 # sai so yeu cau 10D => 1e-11
No = 1000  # so lan lap
p0 = 5*np.random.rand(1)  # Nghiem ban dau random
p1 = p0 + 10
q0 = f(p0)
q1 = f(p1)

for i in range(No):
    p = p1 - q1*(p1 - p0)/(q1 - q0)
    q = f(p)
    if abs(p - p1) < TOL:
        break
    elif q*q1 < 0:
        p0 = p1
        q0 = q1
    p1 = p
    q1 = q

print(p)
```

def get_L(x, x_data, k=0, eps=1e-6):
    numerator = np.cumprod([x - xi for xi in x_data if xi != x_data[k]])[-1]
    denominator = np.cumprod([x_data[k] - xi for xi in x_data if xi != x_data[k]])[-1]
    return numerator/(denominator + eps)

// Noi suy Newton
import numpy as np
from numpy import cos

N_coeffs = []

def get_N_coeff(x, f, deg=0, end_deg=3):
    global N_coeffs
    
    if deg == end_deg:
        return [f[0]]
    N = len(f) - deg if deg >= 1 else len(f) - 1
    N = max(1, N)
    next_f = [(f[i+1] - f[i]) / (x[i+deg+1] - x[i]) for i in range(N)]
    if deg == 0:
        N_coeffs.extend([f[0]])
    N_coeffs.extend([next_f[0]])
    return [f[0], get_N_coeff(x, next_f, deg + 1)[0]]
    
x_data = [8.1,   8.3,   8.6,   8.7]
f =      [16.94, 17.56, 18.50, 18.82]
Newton_deg = 3
get_N_coeff(x_data, f, end_deg=Newton_deg)

x = 8.4
estimate_value = f[0]

for i in range(1, len(N_coeffs)):
    temp = 1
    for j in range(i):
        temp *= (x - x_data[j])
    estimate_value += temp*N_coeffs[i]
print(estimate_value)
